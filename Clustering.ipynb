```python
# Instalar librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
# Importar datasets de iris 
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, :2] 
y = iris.target
```

```python
# Verifico el contenido del dataset
iris
```

```python
# Identificar el número óptimo de clusters por medio del "método del codo"

from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = "k-means++", max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title("Método del codo")
plt.xlabel("Número de Clusters")
plt.ylabel("WCSS(k)")
plt.show()
```

```python
# Aplicar el método de k-means para segmentar el data set

kmeans = KMeans(n_clusters = 3, init="k-means++", max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualización de los clusters

print(y_kmeans)
```

```python
# Visualización de los clusters del dataset de Iris

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 60, c = "red", label = "Grupo 1")
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 60, c = "blue", label = "Grupo 2")
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 60, c = "green", label = "Grupo 3")
plt.title("Cluster del Dataset de Iris")
plt.legend()
plt.show()
```
